**Task 3: Handling Null Values**

	•	Check which columns have missing/null values.
	•	Count the number of null values per column.
	•	Delete rows that have nulls or missing values
	•	Challenge tass: Delete rows that have nulls in Quantity or UnitPrice.

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)

data.isnull().sum()  # we are checking whether nulls have been removed. As we can see it shows 0 for all columns
```

```python
#Quantity or UnitPrice.

data = data.dropna(subset=['Quantity','UnitPrice'])
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

**Task 6: Changing Data Types**

	•	Convert Date to a date/time type.
	•	Convert Quantity to integer (after handling nulls).

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Quantity'] = data['Quantity'].astype(int)
```

```python
data['TotalPrice'] = data['TotalPrice'].astype(float)
```
