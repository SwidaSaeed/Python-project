```python
#Question: "Write the code to create a pivot table with 'class' as rows, 'gender' as columns, and 'mark' as values."


new_piv = data.pivot_table(
    index = 'class',
   columns = 'gender',
    values = 'score',
    aggfunc= 'mean'
).round(2)
new_piv

#Question: "Write the code to create a new column 'grade' where marks >= 85 are 'A', 70-84 are 'B', 60-69 are 'C', and below 60 are 'D'."

import numpy as np
#defined the conditions first
conditions = [
    (data['score']>=85),
    (data['score']>=70) & (data['score']<= 84),
    (data['score']>=60) & (data['score']<= 69),
    (data['score']<60)
]


#Grants example

#df['grade'] = np.where(df['mark'] >= 85, 'A',
#              np.where(df['mark'] >= 70, 'B',
#              np.where(df['mark'] >= 60, 'C', 'D')))
#df

#defining the results

results = ['A','B','C','D']

#creating a new column based on conditions

data['grade'] = np.select(conditions, results,default='')

data


#Question: "Write the code to sort the DataFrame by 'mark' in descending order."

sort_by_scores = data.sort_values(by='score', ascending=False)

sort_by_scores
```
